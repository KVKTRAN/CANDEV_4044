```python
from glob import glob
import numpy as np
import pandas as pd
import folium
from folium import plugins
from folium.plugins import HeatMap

lon, lat = -86.276, 30.935 
zoom_start = 5
```

```python
path = 'AIS\\'
path += 'AIS_2020_01_clean.csv'

df_AIS = pd.read_csv(path)
# df_AIS[df_AIS['MMSI'] == 219265000]
df_AIS
```

```python
data = df_AIS.loc[0:1000]
heat_df = data[['Lat', 'Lon', 'SOG']]
heat_df = heat_df.dropna(axis=0, subset=['Lat','Lon'])

heat_data = [[row['Lat'], row['Lon'], row['SOG']] for index, row in heat_df.iterrows()]
```

```python
map_hooray = folium.Map([heat_data[0][0], heat_data[0][1]], tiles='stamentoner', zoom_start=6)

# map_hooray = folium.Map([heat_data[0][0], heat_data[0][1]], zoom_start=6)
HeatMap(heat_data).add_to(map_hooray)
map_hooray
```

```python
map_hooray.save("index.html")
```
