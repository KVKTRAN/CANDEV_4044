```python
import numpy as np
import pandas as pd

path = 'AIS\\'
path += 'AIS_2020_01_clean.csv'

df_AIS = pd.read_csv(path)
df_AIS[df_AIS['MMSI'] == 219265000]
```

```python
import folium 

lat = df_AIS[df_AIS['MMSI'] == 219265000]['Lat']
lon = df_AIS[df_AIS['MMSI'] == 219265000]['Lon']
# print(lat)
# print(lon)

m = folium.Map(location=[lat[0], lon[0]], zoom_start=13)

folium.Circle(
        radius=100,
        location=[lat[1], lon[1]],
        color="crimson",
        fill=False,
).add_to(m)

for i in range(0, len(lat)):
    folium.Circle(
        radius=100,
        location=[lat[i], lon[i]],
        color="crimson",
        fill=False,
    ).add_to(m)

m
```
